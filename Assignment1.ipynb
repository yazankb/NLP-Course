## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from collections import defaultdict 
import re 
import copy
class BPETokenizer:
    """
    Byte Pair Encoding (BPE) Tokenizer.

    BPE is a simple form of data compression that converts the most frequent pairs of characters in a dataset into single byte representations.
    This tokenizer aims to segment a corpus into subword units, which can help improve the handling of unknown words and rare words in Natural Language Processing tasks.

    Attributes:
    - vocab_size (int): The desired size of the vocabulary after tokenization.
    - operations (list): List to store the merge operations (pairs of characters) used to build the vocabulary.

    Methods:
    - _get_vocab(corpus: List[str]) -> None: Given a list of sentences, returns the frequency of each word in the sentences.
    - _frequent_pair() -> None: Finds the most frequent pair of tokens to be used to update the vocabulary.
    - _update_vocab() -> None: Updates the vocabulary based on the most recent merge operation.
    - train(corpus: List[str]) -> None: Trains the tokenizer on a corpus to build the vocabulary.
    - tokenize(text: str) -> List[str]: Tokenizes a text string based on the learned vocabulary.
    """

    def __init__(self, vocab_size: int = None) -> None:
        """
        Initialize the BPETokenizer.

        Args:
        - vocab_size (int, optional): The desired size of the vocabulary after tokenization. If not provided, it will be determined based on the number of unique characters in the training corpus.
        """
        self.vocab_size = vocab_size  # vocabulary size
        self.current_size = None # initial
        self.operations = []  # Initialize operations list

    def _get_vocab(self, corpus: List[str]) -> None:
        """
        Given a list of sentences, return the frequency of each word in the sentences.

        Args:
        - corpus (List[str]): List of sentences to build the vocabulary from.
        """
        vocab = defaultdict(int)  # Initialize defaultdict to store vocabulary
        for line in corpus:
            for word in line.split():
                word = ' '.join(list(word)) + ' </w>'  # Add </w> to mark end of word
                vocab[word] += 1  # Increment frequency count for each word
        self.vocab = vocab  # Store vocabulary

    def _frequent_pair(self) -> None:
        """
        Finds the most frequent pair of tokens to be used to update the vocabulary.
        """
        pairs = defaultdict(int)  # Initialize defaultdict to store pairs frequencies
        for word, frq in self.vocab.items():
            chars = word.split()
            for i in range(len(chars) - 1):
                pairs[chars[i], chars[i+1]] += frq  # Increment frequency for each pair
        self.operations.append(max(pairs, key=pairs.get))  # Store most frequent pair in operations list

    def _update_vocab(self) -> None:
        """
        Updates the vocabulary based on the most recent merge operation.
        """
        operation = self.operations[-1]  # Get the most recent merge operation
        new_vocab = {}
        bigram = re.escape(' '.join(operation))  # Escape special characters in operation
        p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')  # Compile regex pattern
        for word in self.vocab: 
            w_out = p.sub(''.join(operation), word)  # Substitute operation in each word
            new_vocab[w_out] = self.vocab[word]  # Update frequency in new vocabulary
        self.vocab = new_vocab  # Update vocabulary
        self.current_size += 1  # Increment vocabulary size

    def train(self, corpus: List[str]) -> None:
        """
        Train the tokenizer on a corpus to build the vocabulary.

        Args:
        - corpus (List[str]): List of sentences to train the tokenizer on.
        """
        # No work needed
        if self.vocab_size == None:
            return
        self._get_vocab(corpus)  # Build initial vocabulary
        if self.current_size is None:
            # Determine vocabulary size based on unique characters in the corpus
            unique_chars = set(''.join(self.vocab.keys()))
            self.current_size = len(unique_chars)
        
        while self.current_size < self.vocab_size:
            self._frequent_pair()  # Find frequent pairs
            self._update_vocab()  # Update vocabulary based on pairs

    def tokenize(self, text: str) -> List[str]:
        """
        Tokenize a text string based on the learned vocabulary.

        Args:
        - text (str): Input text to be tokenized.

        Returns:
        - List[str]: List of tokens generated from the input text.
        """
        # Convert the input text into a list of characters
        text = list(text)

        # Replace spaces with '</w>' to mark end of word
        tokens = [symbol.replace(' ', '</w>') for symbol in text]

        new_tokens = []  # List to store updated tokens


        # Iterate through each merge operation
        for operation in self.operations:
            i = 0
            last_token = False  # Flag to indicate if the last token was processed
            # Iterate through the tokens
            while i < (len(tokens) - 1):
                # Check if the current token and the next token form the merge operation
                if tokens[i] == operation[0] and tokens[i+1] == operation[1]:
                    new_tokens.append(tokens[i] + tokens[i+1])  # Merge the tokens
                    if i == len(tokens) - 2:
                        last_token = True  # Mark if the last token was processed
                    i += 1  # Increment index to skip the next token
                else:
                    new_tokens.append(tokens[i])  # Append the current token
                i += 1  # Move to the next token
            if not last_token:
                new_tokens.append(tokens[-1])  # Append the last token if not processed
            tokens = new_tokens.copy()  # Update the tokens with the merged tokens
            new_tokens = []  # Reset the list for updated tokens

        return tokens  # Return list of tokens
```

#### Loading the data

```python
import os

# Define the relative path to the text files
relative_paths = ["assignment_data/file1",
                  "assignment_data/file2",
                  "assignment_data/file3",
                  "assignment_data/file4",
                  "assignment_data/file5"]  # Adjust the path as needed

# Get the absolute path by joining the current directory with the relative path
absolute_paths = [os.path.abspath(relative_path) for relative_path in relative_paths]

contents = []
for path in absolute_paths:
    # Open the text file in read mode with UTF-8 encoding
    with open(path, "r", encoding="utf-8") as file:
        # Read the contents of the file
        file_contents = file.read()
        file_contents = file_contents.split('\n')
        contents.extend(file_contents)

# Print the contents of the file
print(len(contents))
```

#### Analysis

```python
import matplotlib.pyplot as plt

# Define the list of vocab_sizes to experiment with
vocab_sizes = [None, 250, 500, 750, 1000]

# Define a dictionary to store the average length of tokenized tweets for each vocab_size
avg_lengths = {}

# Iterate over each vocab_size
for vocab_size in vocab_sizes:
    # Initialize and train the BPE tokenizer
    tokenizer = BPETokenizer(vocab_size=vocab_size)
    tokenizer.train(contents)  # Assuming 'contents' is the list of tweets
    
    # Tokenize the tweets using the trained tokenizer
    tokenized_tweets = [tokenizer.tokenize(tweet) for tweet in contents if tweet != '']

    
    # Calculate the average length of tokenized tweets
    avg_length = sum(len(tweet) for tweet in tokenized_tweets) / len(tokenized_tweets)
    
    # Store the average length for the current vocab_size
    avg_lengths[vocab_size] = avg_length

# Plot the dependency of average length of tokenized tweets on vocab_size
plt.plot(list(avg_lengths.keys()), list(avg_lengths.values()), marker='o')
plt.xlabel('Vocabulary Size')
plt.ylabel('Average Length of Tokenized Tweets')
plt.title('Dependency of Average Length on Vocabulary Size')
plt.grid(True)
plt.show()
```

#### As the vocabulary size increases, the average length of the tokenized tweets decreases. This trend is expected because, with a larger vocabulary, the Byte Pair Encoding (BPE) algorithm can encode more information per token, resulting in fewer tokens needed to represent the same amount of information. However, it's important to add that there is a trade-off when choosing the size of the vocabulary. A larger vocabulary might reduce the length of the tokenized tweets, but it also requires more memory to store and may lead to increased computational complexity during model training and inference. From the graph we can see that the vocabulary size of 1000 is a good choice

```python
import json

json_filename = 'assignment_data/GermanQuAD_test.json'

# Now let's read from the file and extract the text
def extract_text_from_json(file_path):
    # Read the JSON data from the file
    with open(file_path, 'r', encoding='utf-8') as file:
        json_data = json.load(file)
        
    # Extract the 'context' text from the paragraphs in the data
    extracted_texts = []
    for item in json_data['data']:
        for paragraph in item['paragraphs']:
            extracted_texts.append(paragraph['context'])
            
    return extracted_texts

# Call the function and get the list of extracted texts
extracted_texts = extract_text_from_json(json_filename)
extracted_texts = extracted_texts[:10]
```

```python
import matplotlib.pyplot as plt

vocab_sizes = [None, 250, 500, 750, 1000, 1500, 2000, 2500] 

# Define a dictionary to store the average length of tokenized tweets for each vocab_size
avg_lengths = {}

# Iterate over each vocab_size
for vocab_size in vocab_sizes:
    # Initialize and train the BPE tokenizer
    tokenizer = BPETokenizer(vocab_size=vocab_size)
    tokenizer.train(extracted_texts)  # Assuming 'extracted_texts' is the list of tweets
    
    # Tokenize the tweets using the trained tokenizer
    tokenized_tweets = [tokenizer.tokenize(tweet) for tweet in extracted_texts if tweet is not None]
    
    # Calculate the average length of tokenized tweets
    avg_length = sum(len(tweet) for tweet in tokenized_tweets) / len(tokenized_tweets)
    
    # Store the average length for the current vocab_size
    avg_lengths[vocab_size] = avg_length

# Plot the dependency of average length of tokenized tweets on vocab_size
plt.plot(list(avg_lengths.keys()), list(avg_lengths.values()), marker='o')
plt.xlabel('Vocabulary Size')
plt.ylabel('Average Length of Tokenized Tweets')
plt.title('Dependency of Average Length on Vocabulary Size')
plt.grid(True)
plt.show()
```
